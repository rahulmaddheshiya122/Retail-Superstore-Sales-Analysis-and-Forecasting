```python
#Using Pandasql
#SQL method 1

print(sqldf('''select Customer_ID, Customer_Name,
               cus_avg_order_value,
               RANK() OVER(ORDER BY cus_avg_order_value desc) as Customer_Rank
               from 
               (
               select Customer_ID, Customer_Name,
               avg(Sales) as cus_avg_order_value
               from df2
               group by Customer_ID, Customer_Name
               )
            '''))
```

<h3>Question 5: Give the name of customers who ordered highest and lowest orders from each city.

```python
#Using Pandasql

print(sqldf('''With max_value as(
               Select customer_name as high_value_customer,
               city,
               max(Sales) as high_order_value
               from df2
               group by City),
               
               min_value as(
               Select customer_name as low_value_customer,
               city,
               min(Sales) as low_order_value
               from df2
               group by City)
               
               select high_value_customer, high_order_value, low_value_customer, low_order_value, max_value.city
               from max_value
               inner join
               min_value
               on max_value.city = min_value.city

'''))
```

```python
df2['City'].unique().shape[0]
```

```python
#Using Pandas

high_low_orders = df2.groupby(['City'])['Sales'].agg([max,min])

#Merging with the original df to get the corresponding customer names
result = df2.merge(high_low_orders, on='City')

#Filter rows with the maximum and minimum order values for each city
highest_customers = result[result['Sales'] == result['max']]
lowest_customers = result[result['Sales'] == result['min']]

# Select the columns containing customer names and order values for the highest and lowest orders
highest_customers = highest_customers[['City', 'Customer_Name', 'max']]
lowest_customers = lowest_customers[['City', 'Customer_Name', 'min']]

# Rename the columns to indicate the highest and lowest order values
#highest_customers.rename(columns={'max': 'high_order_value'}, inplace=True)
#lowest_customers.rename(columns={'min': 'low_order_value'}, inplace=True)

# Merge both DataFrames on 'City' to combine the information
result = highest_customers.merge(lowest_customers, on='City', suffixes=('_HighOrder','_LowOrder')).sort_values('City')
result
```

<h3>Question 11: Which segment places the highest number of orders from each state and which segment places the largest individual orders from each state?

```python
#Using Pandasql

#Which segment places the highest number of orders from each state
print(sqldf('''SELECT State, Segment, num_orders 
               from
               (
               SELECT State,Segment,
               count(Order_ID) as num_orders,
               row_number() over(partition by State order by count(Order_ID) desc) as rank_num
               from df2
               group by State, Segment
               )
               where rank_num=1
            '''))
```

```python
#Using Pandasql

#which segment places the largest individual orders from each state
print(sqldf('''SELECT State, Segment, Order_ID, Sales
                FROM (
                  SELECT
                    State,
                    Segment,
                    Order_ID,
                    Sales,
                    ROW_NUMBER() OVER (PARTITION BY State ORDER BY Sales DESC) AS rank_num
                  FROM
                    df2
                ) 
                WHERE rank_num = 1
            '''))
```

<h3>Question 12: Find all the customers who individually ordered on 3 consecutive days where each day’s total order was more than 50 in value.

```python
#Using Pandasql
#method1
print(sqldf('''SELECT distinct
               
                a.Customer_ID, a.Customer_Name, a.Order_Date
                from df2 as a 
                inner join
                df2 as b on a.Customer_ID = b.Customer_ID and a.Order_Date = date(b.Order_Date, '+1 day')
                inner join
                df2 as c on b.Customer_ID = c.Customer_ID and a.Order_Date = date(c.Order_Date, '+2 day') 
                where 
               a.Sales>10 and b.Sales>10 and c.Sales>10
'''))
```

```python
#Using Pandasql
#method2
print(sqldf('''select Customer_ID, Customer_Name, Order_Date, Sales
                from(
                select Customer_ID, Customer_Name, Order_Date,Sales,
               lag(Order_Date, 1) over(partition by Customer_ID order by Order_Date) as pre_order_date,
               lead(Order_Date, 1) over(partition by Customer_ID order by Order_Date) as next_order_date
               from df2
               )
               where 
               Order_Date = Date(pre_order_date, '+1 day') and
               Order_Date = Date(next_order_date, '-1 day') and
               Sales>10
               group by Customer_ID, Customer_Name, Order_Date
'''))
```

<h3> Question 13: Find the maximum number of days for which total sales on each day kept rising.

```python
#Using Pandasql

print(sqldf('''SELECT MAX(Days) AS Maximum_Consecutive_Days
FROM (
  SELECT
    a.Order_Date,
    COUNT(*) AS Days
  FROM
    df2 a
  JOIN
    df2 b ON a.Order_Date <= b.Order_Date
  GROUP BY
    a.Order_Date
  HAVING
    SUM(b.Sales) - SUM(a.Sales) = a.Sales
) AS ConsecutiveDays
'''))
```
